```python
import graphlib

import networkx as nx
import matplotlib.pyplot as plt
```

```python
dependencies = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": ["D", "E"],
    "D": ["F"],
    "E": ["F"],
    "F": [],
}

ts = graphlib.TopologicalSorter(dependencies)
ordered_tasks = list(ts.static_order())
print(ordered_tasks)
# Output: ['F', 'D', 'E', 'B', 'C', 'A']
```

```python
graph = nx.DiGraph()
graph.add_edges_from(
    [("Start", "B"), ("Start", "C"), ("B", "D"), ("C", "D"), ("C", "E"), ("D", "F"), ("E", "F")]
)
pos_spread = nx.spring_layout(graph, k=0.5, scale=2)
plt.tight_layout()
nx.draw_networkx(graph, pos_spread,  arrows=True)
plt.show()
plt.clf()
```

```python
G = nx.DiGraph()
G.add_edges_from([("A", "B"), ("B", "C"), ("A", "C"), ("C", "D")])

pos = nx.spring_layout(G)  # Define node positions
nx.draw_networkx_nodes(G, pos, node_shape='s', node_color='skyblue', node_size=1000) # Squares
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows=True)
plt.show()
```
